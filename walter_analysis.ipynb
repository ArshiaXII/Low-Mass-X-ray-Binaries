### Abstract

<hr style="border:2px solid gray">

**Low Mass X-ray binaries** have been completely neglected in young star clusters (i.e., self-gravitating systems of stars with an age << 100 Myr). This lack of studies is partially motivated by the dearth of observations of X-ray binaries in young star clusters, observations which are difficult to perform.

If the accretor is a **compact object** and the donor a **main sequence star**, the binary can be observed as an X-ray binary:
the matter accreted onto the compact object contributes to the emission of electromagnetic signal, mostly Xâ€“ray.

* soft X-rays, (~0.1 to ~few KeV) from accretion disk: gravitational potential energy of accreted matter converted into radiation 
* hard X-rays, (>few KeV) from a CORONA (Compton emission)
* gamma rays, synchrotron and Compton emission from a jet

**LMXBs features**: 
* donor mass < compact object mass
* Roche lobe : accretion disk, usually Low Mass X-ray binaries

We looked for LMXBs that either fill or partially fill their Roche lobe (R>= 0.8 RL).

**SEVN** stellar population synthesis code is software to carry out stellar population-synthesis, the process of finding a combination of stars likely to be producing the spectrum of a particular galaxy or stellar cluster. Input includes a library of spectra of stars of various types. SEVN model incorporates parameters such as initial mass function and star formation rate to create synthetic populations.

Binary evolution studied via POPULATION SYNTHESIS CODES:
* include models of stellar evolution in a simplified way
* include prescriptions for supernova explosions
* include treatment of binary evolution processes


**Goal of the analysis**:
Assess the impact of different physical parameters on the distribution of LMXBs studying data obtained with the SEVN population-synthesis code.

**Remnant type**

Check the remnant type of the objects in the binary systems of different folders.

Remnants are the final products of the stellar life, they could be a black hole, a neutron star or a white dwarf (not interested in the latter).

Many evolutionary processes can affect the binary, and so the remnant type, such as:

1. wind mass transfer 
2. Roche lobe mass transfer
3. common envelope
4. tidal evolution
5. magnetic braking
6. gravitational wave decay
7. SN kick

Final fate of two stars in a TIGHT binary system very different from that of two single stars or two stars in a loose binary system (semi-major axis >10^5 Rsun ~ 500 AU).

```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

class SeabornFig2Grid():

    def __init__(self, seaborngrid, fig,  subplot_spec):
        self.fig = fig
        self.sg = seaborngrid
        self.subplot = subplot_spec
        if isinstance(self.sg, sns.axisgrid.FacetGrid) or \
            isinstance(self.sg, sns.axisgrid.PairGrid):
            self._movegrid()
        elif isinstance(self.sg, sns.axisgrid.JointGrid):
            self._movejointgrid()
        self._finalize()

    def _movegrid(self):
        """ Move PairGrid or Facetgrid """
        self._resize()
        n = self.sg.axes.shape[0]
        m = self.sg.axes.shape[1]
        self.subgrid = gridspec.GridSpecFromSubplotSpec(n,m, subplot_spec=self.subplot)
        for i in range(n):
            for j in range(m):
                self._moveaxes(self.sg.axes[i,j], self.subgrid[i,j])

    def _movejointgrid(self):
        """ Move Jointgrid """
        h= self.sg.ax_joint.get_position().height
        h2= self.sg.ax_marg_x.get_position().height
        r = int(np.round(h/h2))
        self._resize()
        self.subgrid = gridspec.GridSpecFromSubplotSpec(r+1,r+1, subplot_spec=self.subplot)

        self._moveaxes(self.sg.ax_joint, self.subgrid[1:, :-1])
        self._moveaxes(self.sg.ax_marg_x, self.subgrid[0, :-1])
        self._moveaxes(self.sg.ax_marg_y, self.subgrid[1:, -1])

    
    def _moveaxes(self, ax, gs):
        #https://stackoverflow.com/a/46906599/4124317
        ax.remove()
        ax.figure=self.fig
        self.fig.axes.append(ax)
        self.fig.add_axes(ax)
        ax._subplotspec = gs
        ax.set_position(gs.get_position(self.fig))
        ax.set_subplotspec(gs)

    def _finalize(self):
        plt.close(self.sg.fig)
        self.fig.canvas.mpl_connect("resize_event", self._resize)
        self.fig.canvas.draw()

    def _resize(self, evt=None):
        self.sg.fig.set_size_inches(self.fig.get_size_inches())
```

<hr style="border:2px solid gray">

If a star fills its Roche lobe (radius is equal or larger than Roche lobe), matter flows without energy change to the other star.

**Mass Transfer** called **Roche Lobe Overflow**


Star that fills the RL is called **donor** while the star that receives mass is the **accretor**.

The precise shape of the Roche lobe depends on the mass ratio ${\displaystyle q=M_{1}/M_{2}}{\displaystyle q=M_{1}/M_{2}}$, and must be evaluated numerically.

An approximate formula for the radius of this sphere is:

${\displaystyle {\frac {r_{1}}{A}}=\max {[f_{1},f_{2}]}}, for {\displaystyle 0<q<20}{\displaystyle 0<q<20}$.

The length a is the orbital semi-major axis of the system and $r_1$ is the radius of the sphere whose volume approximates the Roche lobe of mass $M_1$, his formula is accurate to within about 2%.

Another approximate formula was proposed by Eggleton and reads as follows:

${\displaystyle {\frac {r_{1}}{a}}={\frac {0.49q^{2/3}}{0.6q^{2/3}+\ln(1+q^{1/3})}}}$.

This formula gives results up to 1% accuracy over the entire range of the mass ratio ${\displaystyle q}$.
